```python
### Script para mover las imagenes junto a los txt
import os
import shutil

out_path = '../out/'
img_path = '../train_zip/train/'
for element in os.listdir(out_path):
  img_name = element.split('.')[0] + '.jpg'
  img_last = img_path + img_name
  img_new = out_path + img_name
  shutil.copy(img_last, img_new)
```

```python
out_path = '../out_test/'
img_path = '../test_zip/test/'
for element in os.listdir(out_path):
  img_name = element.split('.')[0] + '.jpg'
  img_last = img_path + img_name
  img_new = out_path + img_name
  shutil.copy(img_last, img_new)
```

```python
### Combinar todo en un solo txt
import glob
try:
  os.makedirs('../for_train/')
except:
  print('La carpeta existe')
with open('../for_train/test.txt','w') as f:
  for filepath in glob.iglob('../out_test/*.jpg'):
    img = filepath.split('/')[-1]
    f.write('/content/darknet/out_test/'+img+'\n')
```

```python
with open('../for_train/train.txt','w') as f:
  for filepath in glob.iglob('../out/*.jpg'):
    img = filepath.split('/')[-1]
    f.write('/content/darknet/out/'+img+'\n')
```
