# Tensorflow

```python
import tensorflow as tf
import numpy as np
```

## 1. Tensorflow como numpy

```python
# Tensores
tf.constant([[1., 2., 3., 4.],[1., 6., 2., 8.]])
```

```python
tf.constant(1) # Escalar
```

```python
tensor = tf.constant([[1., 2., 3., 4.],[1., 6., 2., 8.]])
print(tensor.shape)
print(tensor.dtype)
```

Indexar

```python
# Indexar
tensor[:,2:]
```

```python
tensor[tf.newaxis,:,2:]
```

Operaciones matematicas

```python
tensor + 2
```

```python
tf.sqrt(tensor)
```

```python
tensor @ tf.transpose(tensor)
```

## 3. Tensores

```python
a = np.array([1., 2., 4.])
tensor = tf.constant(a)
tensor
```

```python
tensor.numpy()
```

Conversion de tipos

```python
np.array(1) + np.array(3.)
```

```python
tf.constant(1) + tf.constant(3.)
```

```python
tf.constant(1, dtype=tf.float32) + tf.constant(3.)
```

```python
tf.cast(tf.constant(1), tf.float32) + tf.constant(3.)
```

Variables

```python
tensor = tf.constant([1., 2., 4.])
tensor
```

```python
tensor[1] = 5.
```

```python
variable = tf.Variable([1., 2., 4.])
variable
```

```python
variable[1].assign(5.)
```

Estructuras de datos

```python
tf.SparseTensor() # Representa tensores con muchos ceros
tf.TensorArray()
```
